```python
import numpy as np
import pandas as pd
import seaborn
import plotly.express as px
df=pd.read_csv('/content/sample_data/Olympics 2024.csv')
df.head()
k=df.to_excel('Olypicsworld2024.xlsx')
k
```

#extracting medals#


```python
columns_reqd=df[['NOC','Gold','Silver','Bronze','Total']]
#print(columns_reqd)
gold_countries=df[['NOC','Gold']]
Gold_countries=gold_countries.rename(columns={'NOC':"Country",'Gold':"Gold Medal"})
#print(Gold_countries)
Gold_countries=Gold_countries.sort_values(by=['Gold Medal'],ascending=False)
#print(Gold_countries)
y=Gold_countries.head(10)
pieChartForGold=px.pie(y,values='Gold Medal',names='Country',title='countries acquired gold medal top 10 list')
pieChartForGold.show()
```

#competition List#

```python
competition_list=df.groupby('Competitions').size().reset_index(name='Total_count')
print(competition_list)
competition_list.to_excel('competition_list_medalcount.xlsx',index=False)
```
